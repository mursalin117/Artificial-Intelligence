```python
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
import numpy as np
```

```python
# np.random.seed(0)
x = np.linspace(-10, 10, 1000)
# print(x)
# y = 3x³ + 7x² - 12x + 2
y = 3 * (x ** 3) + 7 * (x ** 2) - 12 * x + 2
# print(y)
```

```python
model = Sequential([
    Dense(units=64, activation='relu', input_shape=(1, )),
    Dense(units=64, activation='relu'),
    # Dense(units=64, activation='relu'),
    Dense(units=1)
])

model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')

model.fit(x, y, epochs=500)
```

```python
# from re import X
x_test= np.arange(-3, 4)
print(x_test)
y_test = 3 * (x_test ** 3) + 7 * (x_test ** 2) - 12 * x_test + 2
print(y_test)

out = model.predict(x_test)

for i in range(len(x_test)):
  print("x = %d\t" %x_test[i] + ' y = %d\t' %y_test[i] + ' y_p = %.2f' %out[i][0])
```
